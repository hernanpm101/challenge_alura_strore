### Informe de Ingresos Totales por Tienda

Este informe detalla los ingresos totales generados por cada una de las cuatro tiendas, basado en la suma de los precios de los productos vendidos. Se presenta la distribución de los ingresos por tienda y un resumen del rendimiento general.

```python
import pandas as pd

# Recalcular el ingreso total para cada tienda y almacenarlo en variables separadas
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

# Crear un diccionario con los ingresos totales
ingresos_por_tienda = {
    'Tienda 1': ingreso_total_tienda1,
    'Tienda 2': ingreso_total_tienda2,
    'Tienda 3': ingreso_total_tienda3,
    'Tienda 4': ingreso_total_tienda4
}

print("Ingreso total por tienda:")
for tienda_nombre, ingreso in ingresos_por_tienda.items():
    print(f"  {tienda_nombre}: {ingreso:,.2f}")

# Calcular el ingreso total global
ingreso_total_global = sum(ingresos_por_tienda.values())
print(f"\nIngreso total global de todas las tiendas: {ingreso_total_global:,.2f}")

# Identificar la tienda con mayores y menores ingresos
tienda_mayor_ingreso = max(ingresos_por_tienda, key=ingresos_por_tienda.get)
valor_mayor_ingreso = ingresos_por_tienda[tienda_mayor_ingreso]

tienda_menor_ingreso = min(ingresos_por_tienda, key=ingresos_por_tienda.get)
valor_menor_ingreso = ingresos_por_tienda[tienda_menor_ingreso]

print(f"\nLa tienda con mayor ingreso es: {tienda_mayor_ingreso} con {valor_mayor_ingreso:,.2f}")
print(f"La tienda con menor ingreso es: {tienda_menor_ingreso} con {valor_menor_ingreso:,.2f}")
```

###Informe de las categorias de productos mas y menos vendidas
Se puede apreciar que los muebles son el producto mas vendido y que los articulos para el hogar son los menos vendidos.

```python
def analizar_categorias_ventas(tienda1_df, tienda2_df, tienda3_df, tienda4_df):
    """
    Junta los dataframes de las cuatro tiendas y analiza las categorías
    de productos más y menos vendidas.

    Args:
        tienda1_df (pd.DataFrame): DataFrame de la tienda 1.
        tienda2_df (pd.DataFrame): DataFrame de la tienda 2.
        tienda3_df (pd.DataFrame): DataFrame de la tienda 3.
        tienda4_df (pd.DataFrame): DataFrame de la tienda 4.
    """
    # Concatenar todos los dataframes en uno solo
    todas_las_tiendas_df = pd.concat([tienda1_df, tienda2_df, tienda3_df, tienda4_df])

    # Calcular las ventas por categoría de producto
    ventas_por_categoria_total = todas_las_tiendas_df['Categoría del Producto'].value_counts()

    # Identificar la categoría más vendida
    categoria_mas_vendida = ventas_por_categoria_total.idxmax()
    cantidad_mas_vendida = ventas_por_categoria_total.max()

    # Identificar la categoría menos vendida
    # Usamos .nsmallest(1) para obtener la categoría con el menor conteo
    categoria_menos_vendida = ventas_por_categoria_total.nsmallest(1).index[0]
    cantidad_menos_vendida = ventas_por_categoria_total.nsmallest(1).iloc[0]

    print("--- Informe de Categorías de Productos ---")
    print(f"La categoría de producto MÁS vendida en todas las tiendas es: '{categoria_mas_vendida}' con {cantidad_mas_vendida} ventas.")
    print(f"La categoría de producto MENOS vendida en todas las tiendas es: '{categoria_menos_vendida}' con {cantidad_menos_vendida} ventas.")
    print("\nDetalle de ventas por categoría:")
    print(ventas_por_categoria_total)

# Llamar a la función con los dataframes existentes
analizar_categorias_ventas(tienda, tienda2, tienda3, tienda4)
```

```python
def calcular_calificaciones_promedio(tienda1_df, tienda2_df, tienda3_df, tienda4_df):
    """
    Calcula y muestra la calificación promedio de los clientes para cada tienda.

    Args:
        tienda1_df (pd.DataFrame): DataFrame de la tienda 1.
        tienda2_df (pd.DataFrame): DataFrame de la tienda 2.
        tienda3_df (pd.DataFrame): DataFrame de la tienda 3.
        tienda4_df (pd.DataFrame): DataFrame de la tienda 4.
    """
    promedio_calificacion_tienda1 = tienda1_df['Calificación'].mean()
    promedio_calificacion_tienda2 = tienda2_df['Calificación'].mean()
    promedio_calificacion_tienda3 = tienda3_df['Calificación'].mean()
    promedio_calificacion_tienda4 = tienda4_df['Calificación'].mean()

    print("--- Calificación Promedio de Clientes por Tienda ---")
    print(f"La satisfacción del cliente en la Tienda 1 es: {promedio_calificacion_tienda1:,.2f}")
    print(f"La satisfacción del cliente en la Tienda 2 es: {promedio_calificacion_tienda2:,.2f}")
    print(f"La satisfacción del cliente en la Tienda 3 es: {promedio_calificacion_tienda3:,.2f}")
    print(f"La satisfacción del cliente en la Tienda 4 es: {promedio_calificacion_tienda4:,.2f}")

# Llamar a la función con los dataframes existentes
calcular_calificaciones_promedio(tienda, tienda2, tienda3, tienda4)
```

###Los productos mas y menos vendidos
Mesa de noche con 210 ventas es el producto mas vendido y Celular ABXY con 157 ventas es el menos vendido.

```python
import pandas as pd

def analizar_productos_vendidos(tienda1_df, tienda2_df, tienda3_df, tienda4_df):
    """
    Junta los dataframes de las cuatro tiendas y analiza los productos
    más y menos vendidos.

    Args:
        tienda1_df (pd.DataFrame): DataFrame de la tienda 1.
        tienda2_df (pd.DataFrame): DataFrame de la tienda 2.
        tienda3_df (pd.DataFrame): DataFrame de la tienda 3.
        tienda4_df (pd.DataFrame): DataFrame de la tienda 4.
    """
    # Concatenar todos los dataframes en uno solo
    todas_las_tiendas_df = pd.concat([tienda1_df, tienda2_df, tienda3_df, tienda4_df])

    # Calcular las ventas por producto
    ventas_por_producto_total = todas_las_tiendas_df['Producto'].value_counts()

    # Identificar el producto más vendido
    producto_mas_vendido = ventas_por_producto_total.idxmax()
    cantidad_mas_vendido = ventas_por_producto_total.max()

    # Identificar el producto menos vendido
    # Usamos .nsmallest(1) para obtener el producto con el menor conteo
    producto_menos_vendido = ventas_por_producto_total.nsmallest(1).index[0]
    cantidad_menos_vendido = ventas_por_producto_total.nsmallest(1).iloc[0]

    print("--- Informe de Productos Más y Menos Vendidos ---")
    print(f"El producto MÁS vendido en todas las tiendas es: '{producto_mas_vendido}' con {cantidad_mas_vendido} ventas.")
    print(f"El producto MENOS vendido en todas las tiendas es: '{producto_menos_vendido}' con {cantidad_menos_vendido} ventas.")
    print("\nDetalle de ventas por producto (primeros 5 y últimos 5):")
    print(ventas_por_producto_total.head(5))
    print("...")
    print(ventas_por_producto_total.tail(5))

# Llamar a la función con los dataframes existentes
analizar_productos_vendidos(tienda, tienda2, tienda3, tienda4)
```
